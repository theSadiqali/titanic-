```python
import seaborn as sns 
import matplotlib.pyplot as plt 

# Load dataset 
df = sns.load_dataset('titanic')
print(df.head())
```

```python
print(df[['sex', 'class', 'survived']].head())
```

```python
sns.barplot(x ='sex', y = 'survived', data = df)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate (0 to 1)')
plt.show()
```

```python
sns.barplot(x = 'class', y = 'survived', data = df)
plt.title('Survival rate by passenger class')
plt.show()
```

```python
sns.catplot(x = 'class', y = 'survived', hue = 'sex', kind = 'bar', data = df)
plt.title('survival rate by class and gender')
plt.show()
```

```python
import pandas as pd 

pivot = pd.crosstab([df['class'], df['sex']], df['survived'], normalize = 'index')

#plot heatmap 
sns.heatmap(pivot, annot = True, cmap = 'YlGnBu')
plt.title('Survival Rates Heatmap (by class and gender)')
plt.ylabel('Class and Gender')
plt.xlabel('Survived')
plt.show()
```
